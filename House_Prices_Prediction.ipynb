## Preprocessing Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1=Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),('scaler', StandardScaler())])
cat_pipe1=Pipeline(steps=[('impute', SimpleImputer(strategy="constant")),('ordinal',OrdinalEncoder())])
pre_1=ColumnTransformer([('num',num_pipe1,con),('cat',cat_pipe1,cat)])
```

```python
X_pre=pre_1.fit_transform(X)
X_pre
```

```python
cols=pre_1.get_feature_names_out()
cols
```

```python
X_pre=pd.DataFrame(X_pre, columns=cols)
X_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
model=LinearRegression()
sel=SequentialFeatureSelector(model,direction='forward')
sel_features=sel.fit_transform(X_pre,Y)
sel_cols=sel.get_feature_names_out()
sel_cols
```

```python
imp_cols=[]
for i in sel_cols:
    s=i.split('__')[1]
    imp_cols.append(s)
imp_cols

```

```python
X_sel=X[imp_cols]
X_sel.head()
```

```python
cat_sel,con_sel=catconsep(X_sel)
```

```python
cat_sel
```

```python
con_sel
```

## Final Preprocessing Pipeline

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2=Pipeline(steps=[('impute',SimpleImputer(strategy="mean")),("scaler",StandardScaler())])
cat_pipe2=Pipeline(steps=[('impute',SimpleImputer(strategy='constant')),("ohe",OneHotEncoder(handle_unknown='ignore'))])
pre_2=ColumnTransformer([('num',num_pipe2,con_sel),('cat',cat_pipe2,cat_sel)])

```

```python
X_sel_pre=pre_2.fit_transform(X_sel).toarray()
X_sel_pre
```

```python
final_cols=pre_2.get_feature_names_out()
final_cols
```

```python
X_sel_pre=pd.DataFrame(X_sel_pre, columns=final_cols)
X_sel_pre
```

```python

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_sel_pre,Y,test_size=0.2,random_state=10)
```

```python
xtrain.shape
```

```python
xtest.shape
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
from PM6func import evaluate_model
evaluate_model(xtrain,ytrain,xtest,ytest,model)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
ls=Lasso(alpha=4)
ls.fit(xtrain,ytrain)
```

```python
evaluate_model(xtrain,ytrain,xtest,ytest,ls)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=4)
rr.fit(xtrain,ytrain)
```

```python
evaluate_model(xtrain,ytrain,xtest,ytest,rr)
```

## GridSearchCV

```python
params={'alpha':np.arange(0.1,100,0.1)}
params
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

## For Ridge Regression

```python
gscv=GridSearchCV(rr,param_grid=params,cv=10,scoring="neg_mean_squared_error")
gscv.fit(xtrain,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
New_ridge=gscv.best_estimator_
New_ridge
```

```python
r2_scores=cross_val_score(New_ridge,xtrain,ytrain,cv=10,scoring='r2')
r2_scores
```

```python
r2_cv=r2_scores.mean()
r2_cv
```

```python
evaluate_model(xtrain,ytrain,xtest,ytest,New_ridge)
```

## For Lasso Regression

```python
gscv2=GridSearchCV(ls,param_grid=params,cv=10,scoring="neg_mean_squared_error")
gscv2.fit(xtrain,ytrain)
```

```python
gscv2.best_params_
```

```python
gscv2.best_score_
```

```python
New_lasso=gscv2.best_estimator_
New_lasso
```

```python
rmse_score=cross_val_score(New_lasso,xtrain,ytrain,cv=10,scoring="neg_root_mean_squared_error")
rmse_score
```

```python
rmse_cv=-rmse_score.mean()
rmse_cv
```

```python
r2_scores2=cross_val_score(New_lasso,xtrain,ytrain,cv=10,scoring='r2')
r2_scores2
```

```python
r2_cv2=r2_scores2.mean()
r2_cv2
```

```python
evaluate_model(xtrain,ytrain,xtest,ytest,New_lasso)
```

It is evident that after performing GridSearchCV the Lasso Regression Model gave a slightly better performance compared to Ridge Regression Model in the testing results, so will be choosing this model for further predictions.
